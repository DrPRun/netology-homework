# Задание 1

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
rating = pd.read_csv('ml-latest-small/ratings.csv')
rating.head(50)
```

```python
def category(mov):
    if mov['rating'] <= 2:
        return 'low'
    elif 2 < mov['rating'] <= 4:
        return 'middle'
    elif (mov['rating'] ==4 or mov['rating'] ==4.5) :
        return 'high'
```

```python
rating['category'] = rating.apply(category, axis = 1)
rating.head()
```

# Задание № 2

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
keyword = pd.read_csv('ml-latest-small/keywords.csv')
keyword.head()
```

```python
for i in range(0,10000):
    for key, val in geo_data.items():
        for city in val:
            if city in keyword['keyword'][i].split():
                print(city)
```

```python
def region(kwrd):
    for key, val in geo_data.items():
        for city in val:
            if city in kwrd['keyword'].split():
                return city
    return 'undefined'
```

```python
keyword['region'] = keyword.apply(region, axis=1)
keyword.head()
```
