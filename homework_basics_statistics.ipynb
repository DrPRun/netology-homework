# Задание № 1

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
URL = 'https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv'
res = requests.get(URL)
df = pd.DataFrame({'surgery?': [],'Age':[], 'rectal temperature': [], 'pulse': [], 'respiratory rate' : [], 'temperature of extremities' : [], 'pain' : [], 'outcome' : [] })
soup = BeautifulSoup(res.text)
rows = res.text.split('\n') 
rows[0].split(',')
for row in rows:
    df_str = row.split(',')
    df.loc[len(df)] = [df_str[0], df_str[1], df_str[3], df_str[4], df_str[5], df_str[6], df_str[10], df_str[22] ]
for j in df.columns:
    df[j] = pd.to_numeric(df[j], errors='coerce' )
df.info()
df.head(50)
```

# Задание № 2

```python
print('mean')
for j in df.columns:
    print( j,':', df[j].mean())
print('\n median')
for j in df.columns:
    print( j,':',df[j].median())
print('\n mode')
for j in df.columns:
    print(j,':',df[j].mode())
print('\n std')
for j in df.columns:
    print(j,':',df[j].std())
print('\n var')
for j in df.columns:
    print(j,':',df[j].var())
```

```python
print(df['rectal temperature'].max())
```

```python

for j in df.columns:
    df[j].describe()
    q1 = df[j].quantile(0.25)
    q3 = df[j].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5)*iqr
    upper_bound = q3 + (1.5)*iqr
    remove_outliers = df[df[j].between(lower_bound, upper_bound, inclusive=True)].sort_values(j)
    print(remove_outliers)
# df.info()
```
