```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
URL = 'https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv'
res = requests.get(URL)
df = pd.DataFrame({'surgery?': [],'Age':[], 'rectal temperature': [], 'pulse': [], 'respiratory rate' : [], 'temperature of extremities' : [], 'pain' : [], 'outcome' : [] })
soup = BeautifulSoup(res.text)
rows = res.text.split('\n') 
rows[0].split(',')
for row in rows:
    df_str = row.split(',')
    df.loc[len(df)] = [df_str[0], df_str[1], df_str[3], df_str[4], df_str[5], df_str[6], df_str[10], df_str[22] ]
for j in df.columns:
    df[j] = pd.to_numeric(df[j], errors='coerce' )
df.describe()
```

```python
new_df = pd.read_csv(URL, dtype=np.float64, header = None, na_values=['?'])
new_df.describe()
```

```python
cut_df = new_df[[ 0,1, 3, 4, 5, 6, 10, 22]]
df_nm = cut_df.rename({0 : 'surgery?',1: 'Age', 3 : 'rectal temperature', 4 : 'pulse', 5 : 'respiratory rate', 6 :  'temperature of extremities', 10 : 'pain',22:'outcome'  }, axis = 1)
df_nm.head()
```

```python
# заполним пункт surgery значением 2, если пункт боли был 4 или 5, в противном случае 1
df_nm['surgery?'] = df_nm.apply(
    lambda row: 2 if ((np.isnan(row['surgery?']) and row['pain']) > 3 ) 
    else 1 if (np.isnan(row['surgery?']) and row['pain'] <=3)
    else row['surgery?'],
    axis=1 )
# здесь я уже немного усложнил
df_nm['rectal temperature'] = df_nm.apply(lambda row: 37.8 if (row['temperature of extremities'] == 1  and np.isnan(row['rectal temperature']))  # если температура конечностей в норме, то нормальная температу
                                          else df_nm['rectal temperature'].min() if (row['temperature of extremities'] == 4 and np.isnan(row['rectal temperature'])) # если конечности ледяные, то миним температу
                                          else df_nm['rectal temperature'].quantile(0.25)  if (row['temperature of extremities'] == 3 and np.isnan(row['rectal temperature'])) # если конечности холодные, что квантиль 0.25
                                          else df_nm['rectal temperature'].quantile(0.5) if np.isnan(row['rectal temperature']) # остальные nan заполняем температурой выше нормы
                                          else row['rectal temperature'], # если nan нет, то без изменений
                                          axis = 1)
df_nm.describe()
# в результате такого заполнения увеличилось среднее значение на 0.1, на 0.05 стандартное отклонение, и квантиль 50% на 0.1
```

```python
def pulse_function(mov):
    if np.isnan(mov['pulse']):
        if mov['pain'] == 1:
            return df_nm['pulse'].min()
        elif mov['pain'] == 2:
            return df_nm['pulse'].quantile(0.25)
        elif mov['pain'] == 3:
            return df_nm['pulse'].quantile(0.5)
        elif mov['pain'] == 4:
            return df_nm['pulse'].quantile(0.75)
        else :
            return df_nm['pulse'].quantile(0.99)
    else:
        return mov['pulse']
df_nm['pulse'] = df_nm.apply(pulse_function, axis = 1)
# С таким подходом среднее значение увеличилось на 3, и стандартное отклонение увеличилось на 4
df_nm.describe()
```

```python
def respiratory_function(mov):
    if np.isnan(mov['respiratory rate']):
        if mov['pulse'] < df_nm['pulse'].quantile(0.25) :
            return df_nm['respiratory rate'].quantile(0.15)
        elif  df_nm['pulse'].quantile(0.25) <= mov['pulse'] < df_nm['pulse'].quantile(0.5):
            return df_nm['respiratory rate'].quantile(0.4)
        elif df_nm['pulse'].quantile(0.5) <= mov['pulse'] < df_nm['pulse'].quantile(0.75):
            return df_nm['respiratory rate'].quantile(0.65)
        else:
            return df_nm['respiratory rate'].quantile(0.85)
    else:
        return mov['respiratory rate']
# при таком выборе среднее значение увелится на 0.5, а стандартное отклоненеи уменилось на 1
df_nm['respiratory rate'] = df_nm.apply(respiratory_function, axis = 1)
df_nm.describe()
```

```python
def temp_of_ext(mov):
    if np.isnan(mov['temperature of extremities']):
        if mov['rectal temperature']<= df_nm['rectal temperature'].quantile(0.25):
            return df_nm['temperature of extremities'].quantile(0.25)
        elif  (df_nm['rectal temperature'].quantile(0.25) < mov['rectal temperature']<= df_nm['rectal temperature'].quantile(0.5)):
            return 2
        elif  (df_nm['rectal temperature'].quantile(0.5) < mov['rectal temperature']<= df_nm['rectal temperature'].quantile(0.75)):
            return df_nm['temperature of extremities'].quantile(0.75)
        else:
            return df_nm['temperature of extremities'].max()
    else:
        return mov['temperature of extremities']
df_nm['temperature of extremities']=df_nm.apply(temp_of_ext, axis = 1)
df_nm.describe()
    
    
    
    
    
    
    
    
    
    
```

```python
def pain_col(mov):
    if np.isnan(mov['pain']):
        if mov['pulse'] <= df_nm['pulse'].quantile(0.25):
            return df_nm['pain'].min()
        elif df_nm['pulse'].quantile(0.25) < mov['pulse'] <= df_nm['pulse'].quantile(0.5):
            return df_nm['pain'].quantile(0.25)
        elif df_nm['pulse'].quantile(0.5) < mov['pulse'] <= df_nm['pulse'].quantile(0.75):
            return df_nm['pain'].quantile(0.5)
        elif df_nm['pulse'].quantile(0.75) < mov['pulse'] <= df_nm['pulse'].quantile(0.99):
            return df_nm['pain'].quantile(0.75) 
        else:
            return df_nm['pain'].max() 
    else:
        return mov['pain']
df_nm['pain'] = df_nm.apply(pain_col, axis =1)
df_nm.describe()
    
    
    
    
    
    
    
```
